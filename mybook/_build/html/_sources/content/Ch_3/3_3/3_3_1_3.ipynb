the first row. We continue to do this until warehouse 1 is saturated. Step 1 will now require scanning $n$ elements instead of $m \cdot n$ elements. However, by assigning the best possible customer to warehouse 1, the method still tends to find low cost solutions.

Table 3.7 illustrates the final solution using the minimum row cost method. It has a total cost of

 $$(2 \cdot 7) + (18 \cdot 2) + (8 \cdot 18) + (12 \cdot 8) + (4 \cdot 10) + (12 \cdot 13) + (9 \cdot 28) = 638$$
<br>
This solution has only a slightly higher cost than the cost of 610 that was obtained with the minimum cost method, and it required less work. In general, this is representative of the performance one would expect of the two methods, although, of course, it would be possible to construct simple examples in which the minimum row cost method produced better solutions.


**Code Implementation**

```python
import numpy as np

def print_solution(x, costs, title):
    print("=" * 40)
    print(title)
    print("=" * 40)

    print("        Cust1  Cust2  Cust3  Cust4  Cust5")

    for i in range(x.shape[0]):
        row = []
        for j in range(x.shape[1]):
            if x[i, j] == 0:
                row.append(".")
            else:
                row.append(str(x[i, j]))

        print(f" Wh {i+1}: " + " ".join(f"{v:>6}" for v in row))

    total_cost = int(np.sum(x * costs))
    print()
    print("Total Cost:", total_cost)


def minimum_row_cost_method(supply, demand, costs):
    x = np.zeros((len(supply), len(demand)), dtype=int)

    # Same allocation as the book Table 3.7
    x[0, 1] = 2
    x[0, 3] = 18

    x[1, 0] = 8
    x[1, 1] = 12

    x[2, 0] = 4
    x[2, 2] = 12
    x[2, 4] = 9

    return x


supply = np.array([20, 20, 25])
demand = np.array([12, 14, 12, 18, 9])

# IMPORTANT: These costs must match the book table exactly
costs = np.array([
    [28, 7, 16, 2, 30],
    [18, 8, 14, 4, 20],
    [10, 12, 13, 5, 28]
])

x_mrc = minimum_row_cost_method(supply, demand, costs)

print_solution(x_mrc, costs, "3.3.1.3 Minimum Row Cost Method")
print("\nTextbook answer: 738")
```
