The problem can be formulated as a generalized assignment problem where $x_{ij} = 1$ if customer $i$ is served by warehouse $j$. Every customer must be served by exactly one warehouse, and every warehouse has a capacity constraint on the set of customers it can service.

<br>

$$
\begin{aligned}
\text{minimize} \qquad
& \sum_i \sum_j c_{ij}x_{ij} \\[6pt]
\text{subject to:} \qquad
& \sum_j x_{ij} = 1
\qquad \text{for all customers } i \\[6pt]
& \sum_i d_i x_{ij} \leq b_j
\qquad \text{for all warehouses } j \\[6pt]
& x_{ij} = 0 \text{ or } 1
\end{aligned}
$$

<br>

If we solve the problem as an LP with $(0 \leq x_i \leq 1)$, we get a total cost of 890, but two of the customers are served from two warehouses. This violates the 0–1 constraint on the variables.

&nbsp;&nbsp;&nbsp;&nbsp;We construct a Lagrangian function by penalizing the customer constraints:

$$
\small
L(x,u)
=
\mathop{\operatorname{maximum}}\limits_{u}
\;
\mathop{\operatorname{minimum}}\limits_{x_{ij}\in\{0,1\}}
\;
\sum_i \sum_j c_{ij}x_{ij}
-
\sum_i u_i
\left(
\sum_j x_{ij} - 1
\right)
$$

subject to:

$$
\small
\sum_i d_i x_{ij} \leq b_j
\qquad
\text{for all warehouses } j
$$

or, equivalently:

$$
\small
L(x,u)
=
\mathop{\operatorname{maximum}}\limits_{u}
\;
\mathop{\operatorname{minimum}}\limits_{x_{ij}\in\{0,1\}}
\;
\sum_i \sum_j (c_{ij}+u_i)x_{ij}
-
\sum_i u_i
$$

subject to:

$$
\small
\sum_i d_i x_{ij} \leq b_j
\qquad
\text{for all warehouses } j
$$


Observe that because this problem is a *minimization* in $x$, we construct the Lagrangian as a *maximization* in $u$. When we substitute any fixed value of $u$ into the Lagrangian, this problem becomes a simple multiple knapsack problem. We can treat each warehouse as an independent knapsack problem, and find the least expensive customers for that warehouse. However, because we have dropped the customer constraint, there is no reason why a customer cannot be assigned to more than one warehouse, or in fact, to no warehouse. In particular, if we set $u = 0$ initially, we discover that the optimal solution is $x = 0$. No customers are assigned to any warehouse. To make customers attractive to the warehouse, at least some of the costs must be negative; that is, we must choose initial values for the $u$ vector to be negative enough to make some of the Lagrangian costs negative. We will choose u = (-475, -375, -360, -360)


a somewhat arbitrary choice, but one in which the new Lagrangian costs have at least one negative cost for every customer. We can subtract the second smallest cost in each row to ensure that the smallest cost will be negative. In the first row, we can subtract 475 from each element. Now, every customer is *desired* by at least one warehouse. The new Lagrangian costs are:

This first approximation is already a better bound on the solution than the LP solution, which has a value of 890.

&nbsp;&nbsp;&nbsp;&nbsp;When we now examine the customer constraints, we see that no warehouse took customer 1 or 4, and two warehouses took customer 3. To encourage at least one warehouse to take customers 1 and 4, we want to decrease the cost for those customers; that is, decrease $u_1$ and $u_4$. In order to decrease the number of warehouses that want customer 3, we increase the cost slightly. There are many popular methods for doing this, but they all essentially involve trial and error. We can change all three $u_i$ values at once, or we can change them one at a time. We can take small steps, and keep increasing them until the Lagrangian stops increasing, or we can take large steps too far and then back up. Without elaborating on details, we will briefly illustrate the first couple of steps.

```python
from pyomo.environ import *
import numpy as np

# Textbook data — Section 4.7.6
cost = np.array([[475,  95, 665],
                 [375, 150, 375],
                 [360, 180, 180],
                 [360, 180, 360]], dtype=float)
demand   = np.array([19, 15, 12, 18])
capacity = np.array([18, 27, 20])
n_cust, n_wh = 4, 3

# ── Direct 0-1 IP ────────────────────────────────────────────────────────
model = ConcreteModel()
model.I = RangeSet(1, n_cust)
model.J = RangeSet(1, n_wh)
model.x = Var(model.I, model.J, within=Binary)
model.obj = Objective(
    expr=sum(cost[i-1,j-1]*model.x[i,j] for i in model.I for j in model.J),
    sense=minimize)
def assign_rule(model, i):
    return sum(model.x[i,j] for j in model.J) == 1
model.assign = Constraint(model.I, rule=assign_rule)
def cap_rule(model, j):
    return sum(demand[i-1]*model.x[i,j] for i in model.I) <= capacity[j-1]
model.cap = Constraint(model.J, rule=cap_rule)
SolverFactory('cbc').solve(model)

print('=== 4.7.6 Customer Allocation — Textbook Data ===')
print(f'\nDirect 0-1 IP optimal cost: {int(round(model.obj()))}')
print('Assignment:')
for i in model.I:
    for j in model.J:
        if model.x[i,j].value > 0.5:
            print(f'  Customer {i} → Warehouse {j}  (cost={cost[i-1,j-1]:.0f}, demand={demand[i-1]})')
print('  Textbook answer: optimal cost = 1355')

# ── Lagrangian iterations — textbook trace ───────────────────────────────
def solve_warehouse_knapsack(j, lag_cost):
    """Minimum-cost knapsack for warehouse j given modified costs."""
    mk = ConcreteModel()
    mk.I = RangeSet(1, n_cust)
    mk.y = Var(mk.I, within=Binary)
    mk.obj_k = Objective(expr=sum(lag_cost[i-1,j]*mk.y[i] for i in mk.I), sense=minimize)
    mk.cap_k  = Constraint(expr=sum(demand[i-1]*mk.y[i] for i in mk.I) <= capacity[j])
    SolverFactory('cbc').solve(mk)
    assigned = [i for i in mk.I if mk.y[i].value > 0.5]
    return assigned, mk.obj_k()

# Three u vectors traced in the textbook
u_vectors = [
    np.array([-475., -375., -360., -360.]),   # Iteration 1 (textbook starting point)
    np.array([-675., -375., -160., -560.]),   # Iteration 2 (u changed by 200)
    np.array([-485., -375., -350., -370.]),   # Iteration 3 (u changed by 10)
]

print('\nLagrangian iterations (textbook trace):')
for k, u in enumerate(u_vectors):
    lag_cost = cost + u[:, np.newaxis]   # cij + ui
    print(f'\n  Iteration {k+1}:  u = {u.tolist()}')
    print('  Modified cost matrix:')
    for i in range(n_cust):
        print(f'    Customer {i+1}: {lag_cost[i].tolist()}')
    total = 0
    for j in range(n_wh):
        assigned, obj_k = solve_warehouse_knapsack(j, lag_cost)
        print(f'    Warehouse {j+1}: assigns customers {assigned}, knapsack cost={obj_k:.0f}')
        total += obj_k
    lag_val = total - sum(u)
    print(f'  Lagrangian value = {total:.0f} - ({sum(u):.0f}) = {lag_val:.0f}')
```
