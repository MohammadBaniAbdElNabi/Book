In Section 6.2, we noticed, while computing successively higher powers of the transition probability matrix, that state probabilities tended to stabilize regardless of the initial state of the system. This behavior is typical of ergodic Markov chains and is found in most practical applications. When we say that state probabilities become stable, what we really mean is that as the time parameter n becomes large, $\mathrm{P}^{(\mathrm{n}+1)}$ is essentially the same as $\mathrm{P}^{(\mathrm{n})}$. Furthermore, the rows of $\mathrm{P}^{(\mathrm{n})}$ begin to look identical as n grows large, which means that, independent of the initial state of the system, the probabilities of having evolved into each state after any large number of steps do not change. Mathematically, if P is the one-step transition probability matrix, then the powers of P approach a matrix $[\mathscr{P}]$ where

$$
\mathscr{P}=\lim _{n \rightarrow \infty} \mathrm{P}^{n}
$$

in which

1. Each row of $\mathscr{P}$ is an identical probability vector called $\Pi$, and

$$
\Pi_{\mathrm{j}}=\lim _{\mathrm{n} \rightarrow \infty} \mathrm{p}^{(\mathrm{n})}{ }_{\mathrm{ij}} \text { independent of } \mathrm{i} .
$$

2. All elements of $\Pi$ are positive.
3. $\Pi$ has the property that $\Pi \mathrm{P}=\Pi$; that is,

$$
\Pi_{\mathrm{j}}=\sum_{\mathrm{i}=1}^{\mathrm{N}} \Pi_{\mathrm{i}} \mathrm{p}_{\mathrm{ij}} \text { for all } \mathrm{j}=1,2, \ldots, \mathrm{~N}
$$

4. The elements of $\Pi$ represent a probability distribution and therefore

$$
\sum_{\mathrm{j}=1}^{\mathrm{N}} \Pi_{\mathrm{j}}=1
$$

The vector $\Pi$ is called the steady-state probability vector, and each $\Pi_{\mathrm{j}}$ denotes the steadystate probability that the process is in state j after a large number of steps (long after the effects of the initial state have eroded away).

As we cautioned at the end of Section 6.2, when we establish the steady-state (or longterm or equilibrium) behavior of a system, we are not determining that the dynamic system has finally come to rest in some particular state. On the contrary, transitions continue to occur with exactly the probabilities that governed the transitions early in the chain (this is the stationarity assumption). What the steady state probabilities do tell us can be interpreted in several ways:

- $\Pi_{\mathrm{j}}$ is the probability that, if we inspect the system at any instant (long after the process has begun), we will find the system in state j .
- $\Pi_{\mathrm{j}}$ is the percentage of time the system spends in state j in the long run.
- If the Markov chain models the behavior of a large number of entities that all obey the same transition probabilities, then $\Pi_{\mathrm{j}}$ is the fraction (or proportion) of entities that occupy state j at any given time.

Solving the system of equations $\Pi \cdot \mathrm{P}=\Pi$ is much more satisfactory computationally than raising P to higher and higher powers (see Exercise 6.2). However, to obtain the solutions needed, we must note that the rank of the matrix P is $\mathrm{N}-1$, where N is the number of states in the process. (That is, if we add together any $\mathrm{N}-1$ rows, we get the remaining row.) Therefore, we have a system of N dependent equations in N unknowns, and consequently infinitely many solutions. We want to obtain that unique solution for which the unknowns are a probability distribution; therefore, we discard any one of the first N equations in $\Pi \cdot \mathrm{P}=\Pi$ and introduce the equation

$$
\sum_{j=1}^{N} \Pi_{j}=1
$$

We now have a system of N independent equations in which we can uniquely solve the N unknowns $\Pi_{\mathrm{j}}$ for $\mathrm{j}=1,2, \ldots, \mathrm{~N}$.

To illustrate this, we can re-examine the daily changes in the weather system at the ski resort. This process is ergodic, so we can apply a steady state analysis:

$$
\left[\Pi_{1} \Pi_{2} \Pi_{3}\right]\left[\begin{array}{lll}
0.7 & 0.1 & 0.2 \\
0.2 & 0.7 & 0.1 \\
0.5 & 0.2 & 0.3
\end{array}\right]=\left[\Pi_{1} \Pi_{2} \Pi_{3}\right]
$$

The three dependent equations are:

$$
\begin{aligned}
\Pi_{1} & =0.7 \Pi_{1}+0.2 \Pi_{2}+0.5 \Pi_{3} \\
\Pi_{2} & =0.1 \Pi_{1}+0.7 \Pi_{2}+0.2 \Pi_{3} \\
\Pi_{3} & =0.2 \Pi_{1}+0.1 \Pi_{2}+0.3 \Pi_{3}
\end{aligned}
$$

We can arbitrarily choose to discard any one of the three (how about the first one?), and use instead the normalizing equation $\Sigma \Pi_{\mathrm{j}}=1$ to obtain the system:

$$
\begin{aligned}
& 0.1 \Pi_{1}-0.3 \Pi_{2}+0.2 \Pi_{3}=0 \\
& 0.2 \Pi_{1}+0.1 \Pi_{2}-0.7 \Pi_{3}=0 \\
& \Pi_{1}+\Pi_{2}+\Pi_{3}=1
\end{aligned}
$$

The solution to this system of simultaneous linear equations is

$$
\begin{aligned}
& \Pi_{1}=0.5135135 \\
& \Pi_{2}=0.2972973 \\
& \Pi_{3}=0.1891892
\end{aligned}
$$

If we compare these results with the elements in the higher powers of $P$ that we computed in Section 6.2 (as shown in Table 6.1), we find that indeed the value $\Pi_{1}=0.5135135$ appears in all rows of column 1 , the value $\Pi_{2}=0.2972973$ appears in all rows of column 2, and the value $\Pi_{3}=0.1891892$ appears in all rows of column 3 . However, this pattern does not stabilize until we compute $\mathrm{P}^{\mathrm{n}}$ for n values around 26 or 27. (The pattern becomes apparent at about $\mathrm{P}^{10}$, but small changes continue to be evident as we compute successively higher powers, up to about $\mathrm{P}^{27}$, after which no significant changes in $\mathrm{P}^{\mathrm{n}}$ occur.)

The computational effort required to raise P to the 27-th power is considerably greater than the effort required to solve the system of three equations. Furthermore, we have no way of knowing in advance just exactly what power of P needs to be computed. Exercise 6.2 will allow you to observe this contrast for yourself. Solving the steady-state equations is clearly the preferred method for determining the steady-state probabilities.

**Code Implementation**

```python
import numpy as np
import pandas as pd
!pip install pydtmc
from pydtmc import MarkovChain
from pydtmc.plotting import plot_graph
```

```python
P = np.array([[0.7,0.1,0.2],[0.2,0.7,0.1],[0.5,0.2,0.3]])
states = ['Sunny','Cloudy','Snowing']
N = len(P)

# ── Method 1: Textbook exact linear system (p.265) ───────────────────────
print('=== 6.5 Steady-State Analysis ===')
print('\nTextbook linear system:')
print('  0.1·Π1 - 0.3·Π2 + 0.2·Π3 = 0   (from row 2 of ΠP=Π)')
print('  0.2·Π1 + 0.1·Π2 - 0.7·Π3 = 0   (from row 3 of ΠP=Π)')
print('  Π1  +  Π2  +  Π3           = 1   (normalisation)')

A_tb = np.array([[ 0.1, -0.3,  0.2],
                 [ 0.2,  0.1, -0.7],
                 [ 1.0,  1.0,  1.0]])
b_tb = np.array([0.0, 0.0, 1.0])
Pi_tb = np.linalg.solve(A_tb, b_tb)

print('\nSolution (numpy.linalg.solve — Method 1, textbook system):')
tb_vals = [0.5135135, 0.2972973, 0.1891892]
for j, s in enumerate(states):
    ok = abs(Pi_tb[j]-tb_vals[j]) < 1e-6
    print(f'  Π_{j+1} ({s:<8}) = {Pi_tb[j]:.7f}   Textbook: {tb_vals[j]:.7f}  {"✓" if ok else "✗"}')

# ── Method 2: General formula (P^T - I) ──────────────────────────────────
A_gen = (P.T - np.eye(N))
A_gen[-1, :] = 1.0
b_gen = np.zeros(N); b_gen[-1] = 1.0
Pi_gen = np.linalg.solve(A_gen, b_gen)

print('\nSolution (Method 2 — general (P^T−I) system):')
for j, s in enumerate(states):
    print(f'  Π_{j+1} ({s:<8}) = {Pi_gen[j]:.7f}   '
          f'{"✓" if np.isclose(Pi_gen[j], Pi_tb[j]) else "✗"}')

# ── Method 3: pydtmc (one line) ───────────────────────────────────────────
mc = MarkovChain(P.tolist(), states)
Pi_lib = mc.steady_states[0]
print('\nSolution (Method 3 — pydtmc mc.steady_states):')
for j, s in enumerate(states):
    print(f'  Π_{j+1} ({s:<8}) = {Pi_lib[j]:.7f}   '
          f'{"✓" if np.isclose(Pi_lib[j], Pi_tb[j]) else "✗"}')

# ── Verification ──────────────────────────────────────────────────────────
print('\nVerification Π·P = Π:')
PiP = Pi_gen @ P
for j, s in enumerate(states):
    print(f'  (Π·P)[{j+1}] = {PiP[j]:.7f}   Π[{j+1}] = {Pi_gen[j]:.7f}   '
          f'{"✓" if abs(PiP[j]-Pi_gen[j])<1e-9 else "✗"}')

print('\nInterpretation:')
for j, s in enumerate(states):
    print(f'  In the long run {Pi_gen[j]*100:.2f}% of days are {s}')
```
