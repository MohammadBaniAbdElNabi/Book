# 6.6 EXPECTED FIRST PASSAGE TIMES

[![Open In Colab](https://colab.research.google.com/assets/colab-badge.svg)](https://colab.research.google.com/drive/1maLD6G3MO_CF4lRVmnNjZ8PovrMiO-Li?usp=sharing)

We have defined the first passage time of changing from state $i$ to state $j$ to be the number of transitions made by a Markov process as it goes from state i to state j for the first time. If $\mathrm{i}=\mathrm{j}$, then first passage time is the number of steps before the process returns to the same state, and this is called the first recurrence time. We will denote the first passage time from state i to state j as $\mathrm{T}_{\mathrm{ij}}$.

If the Markov process is certain to go from state i to state j eventually (given that the process ever enters state i ), then $\mathrm{T}_{\mathrm{ij}}$ is a random variable. In Section 6.3, we discussed the first passage probability $\mathrm{f}^{(\mathrm{n})}{ }_{\mathrm{ij}}$, which is the probability that $\mathrm{T}_{\mathrm{ij}}=\mathrm{n}$.

If a process in state i is not certain to ever reach state j , then

$$
\sum_{n=1}^{\infty} f^{(n)}{ }_{i j}<1
$$

Otherwise, the $\mathrm{f}^{(\mathrm{n})}{ }_{\mathrm{ij}}$ are a probability distribution for the first passage times $\mathrm{T}_{\mathrm{ij}}$, and

$$
\sum_{n=1}^{\infty} f^{(n)}{ }_{i j}=1
$$

We can then write the expected first passage times $\mathbf{m}_{\mathbf{i j}}$ from state i to state j as

$$
E\left(T_{i j}\right)=m_{i j}=\sum_{n=1}^{\infty} n f^{(n)}{ }_{i j}
$$

(If $\mathrm{i}=\mathrm{j}$, this is called expected recurrence time.) Using these results, we could answer such questions as:

- How many days might we expect to wait for snowy weather to become sunny?
- After how many days on the average will snowy weather conditions again be snowy, after possibly changing to sunny or cloudy in the meantime?

From a computational standpoint, obtaining expected first passage times using the previous formula is difficult because we have to compute $\mathrm{f}^{(\mathrm{n})}{ }_{\mathrm{ij}}$ for all n . However, in the case of expected recurrence time from state i back to itself, we can simply take the reciprocal of the steady-state probability to obtain

$$
\mathrm{m}_{\mathrm{ii}}=\frac{1}{\Pi_{\mathrm{i}}}
$$

(For example, if a process is in state $\mathrm{i} 1 / 4$ of the time during steady-state, then $\Pi_{\mathrm{i}}=1 / 4$ and $\mathrm{m}_{\mathrm{ii}}=4$. That is, we would expect that an average of four steps are required to return to state i.)

For general i and j , we need a practical way of computing $\mathrm{m}_{\mathrm{ij}}$. Suppose a Markov process is currently in state i . Then with probability $\mathrm{p}_{\mathrm{ij}}$, the process will make a transition to state j for the first time in one step. Otherwise, the first move will be to some state k other than j ; and for each $\mathrm{k}=1, \ldots, \mathrm{~N}, \mathrm{k} \neq \mathrm{j}$, this will happen with probability $\mathrm{p}_{\mathrm{ik}}$. In each of these cases, the first passage time will be 1 (the transition from state i to state k ) plus the expected first passage time from the state k to state j . Therefore,

$$
\mathrm{m}_{\mathrm{ij}}=\left(\mathrm{p}_{\mathrm{ij}}\right)(1)+\sum_{\substack{\mathrm{k}=1 \\ \mathrm{k} \neq \mathrm{j}}}^{\mathrm{N}}\left(\mathrm{p}_{\mathrm{ik}}\right)\left(1+\mathrm{m}_{\mathrm{kj}}\right)
$$

which can be expressed as

$$
m_{i j}=\left(p_{i j}+\sum_{\substack{k=1 \\ k \neq j}}^{N} p_{i k}\right)+\sum_{\substack{k=1 \\ k \neq j}}^{N} p_{i k} m_{k j}
$$

to obtain

$$
\mathrm{m}_{\mathrm{ij}}=1+\sum_{\substack{\mathrm{k}=1 \\ \mathrm{k} \neq \mathrm{j}}}^{\mathrm{N}} \mathrm{p}_{\mathrm{ik}} \mathrm{~m}_{\mathrm{kj}}
$$

Thus, $\mathrm{m}_{\mathrm{ij}}$ is defined in terms of other expected first passage times $\mathrm{m}_{\mathrm{kj}}$. Since the $\mathrm{p}_{\mathrm{ik}}$ are known constants, we simply have a linear equation involving $\mathrm{N}-1$ expected first passage times. However, this formula can be used to obtain an equation describing each of the $\mathrm{m}_{\mathrm{kj}}$ involved in the first equation. The resulting system of $\mathrm{N}-1$ simultaneous linear equations in $\mathrm{N}-1$ unknowns can be solved to obtain unique values for all $\mathrm{N}-1$ expected first passage times into state j.

To answer the question: how many days might we expect to wait before snowy ski conditions change to sunny conditions, we need to compute the expected first passage time

$$
\mathrm{m}_{31}=1+\mathrm{p}_{32} \mathrm{~m}_{21}+\mathrm{p}_{33} \mathrm{~m}_{31}
$$

Since $\mathrm{m}_{31}$ is defined in terms of $\mathrm{m}_{21}$, we also need the equation:

$$
\mathrm{m}_{21}=1+\mathrm{p}_{22} \mathrm{~m}_{21}+\mathrm{p}_{23} \mathrm{~m}_{31}
$$

These two equations can be solved simultaneously to obtain $\mathrm{m}_{31}=2.6316$ and $\mathrm{m}_{21}=4.2105$. Therefore, on the average, it is 2.6316 days before snowy weather first becomes sunny. In the process of finding this result, we also observe that it is an average of 4.2105 days before cloudy conditions become sunny.

If we wish to know the first recurrence time $\mathrm{m}_{22}$ to answer the question: after how many days on the average will cloudy weather again become cloudy?, then we solve

$$
\mathrm{m}_{22}=1+\mathrm{p}_{21} \mathrm{~m}_{12}+\mathrm{p}_{23} \mathrm{~m}_{32}
$$

To do this, we also need

$$
\mathrm{m}_{12}=1+\mathrm{p}_{11} \mathrm{~m}_{12}+\mathrm{p}_{13} \mathrm{~m}_{32}
$$

and

$$
\mathrm{m}_{32}=1+\mathrm{p}_{31} \mathrm{~m}_{12}+\mathrm{p}_{33} \mathrm{~m}_{32}
$$

We solve this system to find that $\mathrm{m}_{12}=8.1818, \mathrm{~m}_{32}=7.2727$, and $\mathrm{m}_{22}=3.3636$. Recall that we can also find $\mathrm{m}_{22}$ more quickly as $1 / \Pi_{2}=1 / .2972973=3.3636$, if we have already computed the steady-state probabilities. Thus, cloudy conditions that change to sunny or snowy can be expected to return to cloudy after 3.3636 days.

**Code Implementation**

```python
import numpy as np
import pandas as pd
!pip install pydtmc
from pydtmc import MarkovChain
from pydtmc.plotting import plot_graph
```

```python
P = np.array([[0.7,0.1,0.2],[0.2,0.7,0.1],[0.5,0.2,0.3]])
states = ['Sunny','Cloudy','Snowing']
N = len(P)
Pi = np.array([0.5135135, 0.2972973, 0.1891892])

# ── Method 1: Textbook explicit system for m_i1 (p.267) ─────────────────
print('=== 6.6 Expected First Passage Times (p.267) ===')
print('\nTextbook system for m_31 and m_21 (→ Sunny):')
print(f'  (1-p33)·m31 - p32·m21 = 1  →  {1-P[2,2]:.1f}·m31 - {P[2,1]:.1f}·m21 = 1')
print(f'  -p23·m31 + (1-p22)·m21 = 1  → -{P[1,2]:.1f}·m31 + {1-P[1,1]:.1f}·m21 = 1')

A_m31 = np.array([[1-P[2,2], -P[2,1]],
                  [-P[1,2],  1-P[1,1]]])
m31, m21 = np.linalg.solve(A_m31, [1.0, 1.0])
print(f'\n  m31 (Snowing→Sunny) = {m31:.4f}   Textbook: 2.6316  '
      f'{"✓" if abs(m31-2.6316)<0.001 else "✗"}')
print(f'  m21 (Cloudy→Sunny)  = {m21:.4f}   Textbook: 4.2105  '
      f'{"✓" if abs(m21-4.2105)<0.001 else "✗"}')

# ── Method 2: General linear system for full M matrix ────────────────────
def mfpt_matrix(P, Pi):
    N = len(P)
    M = np.zeros((N,N))
    for j in range(N):
        non_j = [i for i in range(N) if i!=j]
        A = np.eye(len(non_j))
        for row, i in enumerate(non_j):
            for col, k in enumerate(non_j):
                A[row,col] -= P[i,k]
        m_vals = np.linalg.solve(A, np.ones(len(non_j)))
        for idx, i in enumerate(non_j):
            M[i,j] = m_vals[idx]
        M[j,j] = 1.0/Pi[j]
    return M

M = mfpt_matrix(P, Pi)
M_df = pd.DataFrame(M, index=[f'From {s}' for s in states],
                    columns=[f'To {s}' for s in states])
print('\nFull Expected First Passage Time Matrix M (algebraic):')
print(M_df.round(4).to_string())
print('  (Diagonal = recurrence times = 1/Π)')
print(f'\n  m12={M[0,1]:.4f}  m32={M[2,1]:.4f}  m22={M[1,1]:.4f}')
print(f'  Textbook: 8.1818, 7.2727, 3.3636  '
      f'{"✓" if abs(M[0,1]-8.1818)<0.001 else "✗"}')

# ── Method 3: pydtmc ─────────────────────────────────────────────────────
mc = MarkovChain(P.tolist(), states)
mrt = mc.mean_recurrence_times()
print('\nMean recurrence times via pydtmc:')
for i,s in enumerate(states):
    print(f'  m_{s[:2]}{s[:2]} = {mrt[i]:.4f}   (= 1/Π = {1/Pi[i]:.4f})')

print('\nMean first passage times via pydtmc:')
M_lib = np.zeros((N,N))
for j,sj in enumerate(states):
    fpt = mc.mean_first_passage_times_to(sj)
    for i in range(N):
        M_lib[i,j] = fpt[i] if i!=j else mrt[j]
M_lib_df = pd.DataFrame(M_lib,
    index=[f'From {s}' for s in states],
    columns=[f'To {s}' for s in states])
print(M_lib_df.round(4).to_string())
print(f'\n  All match algebraic solution: '
      f'{"✓" if np.allclose(M, M_lib, atol=1e-3) else "✗"}')
```
