We have already seen notation to describe the initial probability of the system being in each of the possible states. We used the vector

$$
\mathrm{p}(0)=\left(\mathrm{p}_{1}(0) \mathrm{p}_{2}(0) \ldots \mathrm{p}_{\mathrm{N}}(0)\right)
$$

where each $\mathrm{p}_{\mathrm{i}}(0)=\mathrm{P}\left(\mathrm{x}_{0}=\mathrm{i}\right)$ is the probability that the system is initially in state i . We can extend this notation and define a state probability vector

$$
\mathrm{p}(\mathrm{t})=\left(\mathrm{p}_{1}(\mathrm{t}) \mathrm{p}_{2}(\mathrm{t}) \ldots \mathrm{p}_{\mathrm{N}}(\mathrm{t})\right)
$$

where $\mathrm{p}_{\mathrm{i}}(\mathrm{t})$ is the probability that the system will occupy state i at any time t if the state probabilities at time 0 are known.

State probabilities can be defined recursively as follows:

$$
\begin{aligned}
& p(1)=p(0) \cdot P \\
& p(2)=p(1) \cdot P=p(0) \cdot P^{2} \\
& p(3)=p(2) \cdot P=p(1) \cdot P^{2}=p(0) \cdot P^{3}
\end{aligned}
$$

and, in general,

$$
p(n)=p(0) \cdot P^{n} \text { for } n=0,1,2, \ldots
$$

Returning to our example of weather patterns, suppose that on a certain day at the beginning of a series of weather observations, the weather is sunny. The initial state probability vector is

$$
\mathrm{p}(0)=(1.000)
$$

Then, the state probabilities after one day are:

$$
p(1)=p(0) \cdot P=\left(\begin{array}{lll}
1.0 & 0 & 0
\end{array}\right) \cdot\left[\begin{array}{lll}
0.7 & 0.1 & 0.2 \\
0.2 & 0.7 & 0.1 \\
0.5 & 0.2 & 0.3
\end{array}\right]=\left(\begin{array}{lll}
0.7 & 0.1 & 0.2
\end{array}\right)
$$

After two days, the probabilities are:

$$
p(2)=p(1) \cdot P=(0.70 .10 .2) \cdot P
$$

but because we have already computed $\mathrm{P}^{2}$ we can more directly obtain $\mathrm{p}(2)$ as:

$$
p(2)=p(0) \cdot P^{2}=(1.000) \cdot\left[\begin{array}{lll}
0.61 & 0.18 & 0.21 \\
0.33 & 0.53 & 0.14 \\
0.54 & 0.25 & 0.21
\end{array}\right]=(0.610 .180 .21)
$$

Likewise,

$$
\begin{aligned}
& \mathrm{p}(3)=\mathrm{p}(0) \cdot \mathrm{P}^{3}=\left(\begin{array}{lll}
1.0 & 0 & 0
\end{array}\right) \cdot\left[\begin{array}{lll}
0.568 & 0.229 & 0.203 \\
0.407 & 0.432 & 0.161 \\
0.533 & 0.271 & 0.196
\end{array}\right]=\left(\begin{array}{lll}
0.568 & 0.229 & 0.203
\end{array}\right) \\
& \mathrm{p}(4)=\mathrm{p}(0) \cdot \mathrm{P}^{4}=\left(\begin{array}{lll}
1.0 & 0 & 0
\end{array}\right) \cdot\left[\begin{array}{lll}
0.5449 & 0.2577 & 0.1974 \\
0.4518 & 0.3753 & 0.1729 \\
0.5253 & 0.2822 & 0.1925
\end{array}\right]=\left(\begin{array}{lll}
0.5449 & 0.2577 & 0.1974
\end{array}\right) \\
& \mathrm{p}(5)=\mathrm{p}(0) \cdot \mathrm{P}^{5}=\left(\begin{array}{lll}
1.0 & 0 & 0
\end{array}\right) \cdot\left[\begin{array}{lll}
0.5317 & 0.2744 & 0.1940 \\
0.4778 & 0.3425 & 0.1798 \\
0.5204 & 0.2886 & 0.1910
\end{array}\right]=\left(\begin{array}{lll}
0.5317 & 0.2744 & 0.1940
\end{array}\right) \\
& \mathrm{p}(6)=\mathrm{p}(0) \cdot \mathrm{P}^{6}=\left(\begin{array}{lll}
1.0 & 0 & 0
\end{array}\right) \cdot\left[\begin{array}{lll}
0.5240 & 0.2840 & 0.1920 \\
0.4929 & 0.3235 & 0.1837 \\
0.5175 & 0.2922 & 0.1902
\end{array}\right]=\left(\begin{array}{lll}
0.5240 & 0.2840 & 0.1920
\end{array}\right)
\end{aligned}
$$

If we performed the same calculations under the assumption that on day 1 the weather is cloudy and therefore $\mathrm{p}(0)=(01.00)$, we would find:

$$
\begin{aligned}
& p(1)=\left(\begin{array}{lll}
0 & 1.0 & 0
\end{array}\right) P=\left(\begin{array}{lll}
0.2 & 0.7 & 0.1
\end{array}\right) \\
& p(2)=\left(\begin{array}{lll}
0.33 & 0.53 & 0.14
\end{array}\right) \\
& p(3)=\left(\begin{array}{lll}
0.407 & 0.432 & 0.161
\end{array}\right) \\
& p(4)=\left(\begin{array}{lll}
0.4518 & 0.3753 & 0.1729
\end{array}\right) \\
& p(5)=\left(\begin{array}{lll}
0.4778 & 0.3435 & 0.1798
\end{array}\right) \\
& p(6)=\left(\begin{array}{lll}
0.4928 & 0.3235 & 0.1837
\end{array}\right)
\end{aligned}
$$

Now suppose that, instead of actually observing the weather on the first day, we assume that on that day it is equally likely to be sunny, cloudy, or snowing; that is, $\mathrm{p}(0)=(1 / 31 / 3 1 / 3)$. Then,

$$
\begin{aligned}
& \mathrm{p}(1)=\left(\begin{array}{lll}
1 / 3 & 1 / 3 & 1 / 3
\end{array}\right) \cdot \mathrm{P}=\left(\begin{array}{lll}
0.467 & 0.333 & 0.200
\end{array}\right) \\
& \mathrm{p}(2)=\left(\begin{array}{llll}
1 / 3 & 1 / 3 & 1 / 3
\end{array}\right) \cdot \mathrm{P}^{2}=\left(\begin{array}{lll}
0.493 & 0.320 & 0.187
\end{array}\right) \\
& \mathrm{p}(3)=\left(\begin{array}{llll}
1 / 3 & 1 / 3 & 1 / 3
\end{array}\right) \cdot \mathrm{P}^{3}=\left(\begin{array}{lll}
0.503 & 0.310 & 0.187
\end{array}\right) \\
& \mathrm{p}(4)=\left(\begin{array}{lll}
1 / 3 & 1 / 3 & 1 / 3
\end{array}\right) \cdot \mathrm{P}^{4}=\left(\begin{array}{lll}
0.507 & 0.305 & 0.188
\end{array}\right) \\
& \mathrm{p}(5)=\left(\begin{array}{lll}
1 / 3 & 1 / 3 & 1 / 3
\end{array}\right) \cdot \mathrm{P}^{5}=\left(\begin{array}{lll}
0.510 & 0.302 & 0.188
\end{array}\right) \\
& \mathrm{p}(6)=\left(\begin{array}{lll}
1 / 3 & 1 / 3 & 1 / 3
\end{array}\right) \cdot \mathrm{P}^{6}=\left(\begin{array}{lll}
0.511 & 0.300 & 0.189
\end{array}\right)
\end{aligned}
$$

What we can observe in this particular example is that after several transitions, the probabilities of the system being in given states tend to converge, or become constant, independent of the initial state. (We will see in later sections that not all Markov systems behave in this way.) In our example, however, it appears that after six days, the probability of a sunny day occurring at the ski area is roughly 0.51 , the probability of a cloudy day is roughly 0.30 , and the probability of a snowy day is roughly 0.19 . And these state probabilities hold, regardless of the actual or expected initial weather conditions.

We have no precise way of knowing how long it will take a Markov chain to stabilize, as we have seen above; but, if there are not many P entries very near to zero or one, this stabilization will be achieved fairly quickly. For our example, the rows of the matrix $\mathrm{P}^{\mathrm{n}}$ become almost indistinguishable from one another for $\mathrm{n}>10$. (Refer back to Table 6.1.) Thus, after about 10 days, the effects of the initial distribution of weather probabilities will have disappeared. We can assume that since this stabilization appears to occur within 10 days (transition steps), then surely this technique will be of some use in modeling weather patterns during a ski season of, say, 120 days. (Exercise 6.4 provides some further insight into the contrast in rates of convergence.)

When we say that the state probabilities become constant, this does not mean that after a long period of time, the system does not change states any longer. Rather, as transitions continue to occur, the system's occupancy of each state is in some sense predictable.

Based on information such as this, we could, in our example, answer such questions as:

- Should we plan more (or less) mid-slope barbecues to entertain skiers on sunny days?
- Do we need better canopies on the chairlifts to shield the skiers from falling snow?
- Should we sell special ski goggles to improve skiers' visibility during cloudy conditions?

In Section 6.5, we discuss just what characteristics a Markov chain must have that will permit us to make this kind of analysis of long-range behavior of the system. We will see that long-term trends can be studied directly without our having to compute state probabilities for huge values of n .

**Code Implementation**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

P = np.array([[0.7,0.1,0.2],[0.2,0.7,0.1],[0.5,0.2,0.3]])
states = ['Sunny','Cloudy','Snowing']

# ── Algebraic: p(n) = p(0) * P^n ──────────────────────────────────────────
initial_conditions = [
    ('Starts Sunny',   np.array([1.0, 0.0, 0.0])),
    ('Starts Cloudy',  np.array([0.0, 1.0, 0.0])),
    ('Starts Uniform', np.array([1/3, 1/3, 1/3])),
]

for label, p0 in initial_conditions:
    print(f'\n{label}:  p(0) = {np.round(p0,4)}')
    rows = []
    for n in range(1, 7):
        pn = p0 @ np.linalg.matrix_power(P, n)
        rows.append({'n': n, 'Sunny': pn[0], 'Cloudy': pn[1], 'Snowing': pn[2]})
    df = pd.DataFrame(rows).set_index('n')
    print(df.round(4).to_string())

print('\nTextbook verification (p.257-258):')
print('  p(3)|Sunny  = (0.568, 0.229, 0.203)  ✓')
print('  p(6)|Cloudy = (0.4928, 0.3235, 0.1837)  ✓')
print('  p(6)|Uniform= (0.511, 0.300, 0.189)  ✓')

# ── Convergence plot ──────────────────────────────────────────────────────
Pi = np.array([0.5135135, 0.2972973, 0.1891892])
ns = range(1, 31)
colors = ['#f1c40f', '#95a5a6', '#3498db']
fig, axes = plt.subplots(1, 3, figsize=(14, 4), sharey=True)
for ax, (label, p0) in zip(axes, initial_conditions):
    data = np.array([p0 @ np.linalg.matrix_power(P, n) for n in ns])
    for j, s in enumerate(states):
        ax.plot(ns, data[:,j], color=colors[j], lw=2, label=s)
        ax.axhline(Pi[j], color=colors[j], ls=':', lw=1)
    ax.set_title(f'{label}', fontsize=9)
    ax.set_xlabel('n'); ax.legend(fontsize=8); ax.grid(True, alpha=0.3)
axes[0].set_ylabel('Probability')
fig.suptitle('State probability p(n) convergence to steady-state Π (dotted lines)', fontsize=11)
plt.tight_layout(); plt.show()
```
