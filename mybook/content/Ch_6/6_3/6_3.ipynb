In a typical Markov chain, we frequently observe that states are left and re-entered again and again. We have developed a means of computing the probability $\mathrm{p}^{(\mathrm{n})}{ }_{\mathrm{ij}}$ that a system will leave state i and be in state j after n transitions. But this does not give us any information about whether the system entered state j at any time before the n -th transition. Suppose we are specifically interested in the probability that a system leaves state $i$ and enters state j for the first time after n steps. This is called the first passage probability, and it is clearly related to the n-step probability. However, we must exclude all the ways in which the system may have entered state j before the n -th transition. For example, if we know the probability of a cloudy day on the slopes being followed three days later by a sunny day, we realize that this three-step transition can occur in several ways:

$$
\text{cloudy → cloudy → cloudy → sunny} \\
\text{cloudy → cloudy → sunny → sunny} \\
\text{cloudy → sunny → sunny → sunny} \\
\text{cloudy → cloudy → snowy → sunny} \\
\text{cloudy → snowy → snowy → sunny} \\
\text{cloudy → snowy → sunny → sunny} \\
\text{cloudy → sunny → snowy → sunny} \\
\text{cloudy → sunny → cloudy → sunny} \\
\text{cloudy → snowy → cloudy → sunny} \\
$$

The probability $\mathrm{p}^{(3)}{ }_{21}$ accounts for all of these possible paths. By contrast, the first passage probability will account for only those paths in which a sunny day does not occur until the third step. Thus, we want to measure the probability that one of the following paths will occur:

$$
\text{cloudy → cloudy → cloudy → sunny}\\
\text{cloudy → cloudy → snowy → sunny}\\
\text{cloudy → snowy → snowy → sunny}\\
\text{cloudy → snowy → cloudy → sunny}\\
$$

The first passage probability $\mathrm{f}^{(\mathrm{n})}{ }_{\mathrm{ij}}$ is computed as $\mathrm{p}^{(\mathrm{n})}{ }_{\mathrm{ij}}$ minus the probabilities of the n -step paths in which state j occurs before the n -th step. In one transition, $\mathrm{p}^{(1)}{ }_{\mathrm{ij}}=\mathrm{f}^{(1)}{ }_{\mathrm{ij}}$. Then for larger n ,

$$
\begin{aligned}
& f^{(2)}{ }_{i j}=p^{(2)}{ }_{i j}-f^{(1)}{ }_{i j} p^{(1)}{ }_{j j} \\
& f^{(3)}{ }_{i j}=p^{(3)}{ }_{i j}-f^{(1)}{ }_{i j} p^{(2)}{ }_{j j}-f^{(2)}{ }_{i j} p^{(1)}{ }_{j j} \\
& \cdot \\
& \cdot \\
& f^{(n)}{ }_{i j}=p^{(n)}{ }_{i j}-f^{(1)}{ }_{i j} p^{(n-1)}{ }_{j j}-f^{(2)}{ }_{i j} p^{(n-2)}{ }_{j j}-\ldots-f^{(n-1)}{ }_{i j} p^{(1)}{ }_{j j}
\end{aligned}
$$

or, more succinctly,

$$
f^{(n)}{ }_{i j}=p^{(n)}{ }_{i j}-\sum_{k=1}^{n-1} f^{(k)}{ }_{i j} p^{(n-k)}{ }_{j j}
$$

Returning to our example, we will illustrate $\mathrm{f}^{(\mathrm{n})}{ }_{\mathrm{ij}}$ for $\mathrm{n}=1,2$, and 3 .

$$
\begin{gathered}
\mathrm{F}^{(1)}=\mathrm{P}=\left[\begin{array}{ccc}
0.7 & 0.1 & 0.2 \\
0.2 & 0.7 & 0.1 \\
0.5 & 0.2 & 0.3
\end{array}\right] \\
\mathrm{F}^{(2)}=\left[\begin{array}{ccc}
0.61-(0.7)(0.7) & 0.18-(0.1)(0.7) & 0.21-(0.2)(0.3) \\
0.33-(0.2)(0.7) & 0.53-(0.7)(0.7) & 0.14-(0.1)(0.3) \\
0.54-(0.5)(0.7) & 0.25-(0.2)(0.7) & 0.21-(0.3)(0.3)
\end{array}\right]=\left[\begin{array}{ccc}
0.12 & 0.11 & 0.15 \\
0.19 & 0.04 & 0.11 \\
0.19 & 0.11 & 0.12
\end{array}\right]
\end{gathered}
$$

$$
\begin{aligned}
\mathrm{F}^{(3)} & =\left[\begin{array}{l}
0.568-(0.7)(0.61)-(0.12)(0.7) \\
0.407-(0.2)(0.61)-(0.19)(0.7) \\
0.533-(0.5)(0.61)-(0.19)(0.7)
\end{array}\right]\left[\begin{array}{l}
0.229-(0.1)(0.53)-(0.11)(0.7) \\
0.432-(0.7)(0.53)-(0.04)(0.7) \\
0.271-(0.2)(0.53)-(0.11)(0.7)
\end{array}\right]\left[\begin{array}{l}
0.203-(0.2)(0.21)-(0.15)(0.3) \\
0.161-(0.1)(0.21)-(0.11)(0.3) \\
0.196-(0.3)(0.21)-(0.12)(0.3)
\end{array}\right] \\
& =\left[\begin{array}{lll}
0.057 & 0.099 & 0.116 \\
0.152 & 0.033 & 0.107 \\
0.095 & 0.088 & 0.097
\end{array}\right]
\end{aligned}
$$

And, just to check our intuition, let us re-examine $f^{(3)}{ }_{21}$. This should be $f^{(3)}{ }_{21}=p^{(3)}{ }_{21}$

$$
\begin{aligned}
& \text { - Probability (cloudy → cloudy } \text { → sunny } \text { → sunny) } \\
& \text { - Probability (cloudy → sunny } \text { sunny } \text { sunny) } \\
& \text { - Probability (cloudy → snowy } \text { → sunny } \text { → sunny) } \\
& \text { - Probability (cloudy → sunny } \text { snowy } \text { sunny) } \\
& \text { - Probability (cloudy } \text { sunny } \text { s cloudy } \text { sunny) } \\
& \text { = } 0.407-0.098-0.098-0.035-0.020-0.004 \\
& =0.152
\end{aligned}
$$

which is exactly the value we computed earlier as the element $\mathrm{f}^{(3)}{ }_{21}$ in the matrix $\mathrm{F}^{(3)}$.

**Code Implementation**

```python
import numpy as np
import pandas as pd
!pip install pydtmc
from pydtmc import MarkovChain
```

```python
P = np.array([[0.7,0.1,0.2],[0.2,0.7,0.1],[0.5,0.2,0.3]])
states = ['Sunny','Cloudy','Snowing']
N = len(P)

# ── Algebraic: recursive formula from textbook p.260 ─────────────────────
def first_passage_matrices(P, max_n):
    Pn = [np.eye(N)] + [None]*max_n
    Fn = [None]*(max_n+1)
    for k in range(1, max_n+1):
        Pn[k] = Pn[k-1] @ P
    Fn[1] = Pn[1].copy()
    for n in range(2, max_n+1):
        Fn[n] = Pn[n].copy()
        for k in range(1, n):
            for i in range(N):
                for j in range(N):
                    Fn[n][i,j] -= Fn[k][i,j] * Pn[n-k][j,j]
    return Fn

Fn = first_passage_matrices(P, 3)

print('=== First Passage Probability Matrices (pp.260-261) ===')
for n in [1, 2, 3]:
    df = pd.DataFrame(Fn[n], index=states, columns=states)
    print(f'\nF({n}) — algebraic formula:')
    print(df.round(4).to_string())

print('\nVerification:')
print(f'  f(2)[Sunny,Sunny]  = {Fn[2][0,0]:.4f}   Textbook: 0.12  '
      f'{"✓" if abs(Fn[2][0,0]-0.12)<0.001 else "✗"}')
print(f'  f(3)[Cloudy,Sunny] = {Fn[3][1,0]:.4f}   Textbook: 0.152 '
      f'{"✓" if abs(Fn[3][1,0]-0.152)<0.001 else "✗"}')

print('\nTextbook manual check for f(3)_21 (p.261):')
manual = 0.407 - 0.098 - 0.098 - 0.035 - 0.020 - 0.004
print(f'  0.407 - 0.098 - 0.098 - 0.035 - 0.020 - 0.004 = {manual:.3f}  ✓')

# ── pydtmc: first_passage_probabilities ───────────────────────────────────
print('\n=== pydtmc first_passage_probabilities ===')
mc = MarkovChain(P.tolist(), states)
for j, sj in enumerate(states):
    Fn_lib = mc.first_passage_probabilities(3, sj)
    print(f'\nFirst passage probabilities to {sj}:')
    for n in [1,2,3]:
        row = {si: round(Fn_lib[n-1][i],4) for i,si in enumerate(states)}
        print(f'  F({n}): {row}')
```
