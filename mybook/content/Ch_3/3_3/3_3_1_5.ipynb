I. Compute the reduced costs c′<sub>ij</sub> such that every basic cell has a zero reduced cost.  
&emsp;&emsp;Initially, assume c′<sub>ij</sub> = c<sub>ij</sub>, and the u<sub>i</sub> and v<sub>j</sub> are all zero.

&emsp;&emsp;a. Construct the basic variable network, tree, as in **Figure 3.9**. Select any u<sub>i</sub> and assign to it any arbitrary fixed value.

&emsp;&emsp;b. For each unfixed v<sub>j</sub> that is adjacent to a fixed u<sub>i</sub>, adjust v<sub>j</sub> such that c′<sub>ij</sub> is zero, and then call v<sub>j</sub> fixed.

&emsp;&emsp;c. For each unfixed u<sub>i</sub> that is adjacent to a fixed v<sub>j</sub>, adjust u<sub>i</sub> such that c′<sub>ij</sub> is zero, and call u<sub>i</sub> fixed.

&emsp;&emsp;d. Repeat steps 2 and 3 until all u<sub>i</sub> and v<sub>j</sub> are fixed.

&emsp;&emsp;e. Compute all non-basic costs as c′<sub>ij</sub> = c<sub>ij</sub> − u<sub>i</sub> − v<sub>j</sub>.

II. If any non-basic c′<sub>ij</sub> is negative, let x<sub>ij</sub> enter the basis. As in the ordinary Simplex method, we can choose any negative c′<sub>ij</sub>.

&emsp;&emsp;a. Identify the unique even cycle defined by the edge x<sub>ij</sub> and other basic variable edges.

&emsp;&emsp;b. Alternately increase and decrease the flow in the edges in this circuit until at least one basic variable has a zero flow. Remove that variable from the basis.

&emsp;&emsp;c. Repeat the algorithm completely from the beginning, Part I, by recomputing the reduced costs.

&emsp;&emsp;Continuing with our example, in **Table 3.11**, for Part II of the algorithm, we find c′<sub>34</sub> = −6. Therefore, x<sub>34</sub> can enter the basis. The unique basic cycle is (x<sub>34</sub>, x<sub>35</sub>, x<sub>25</sub>, x<sub>22</sub>, x<sub>12</sub>, x<sub>24</sub>). The increase of the flow in this alternating circuit is limited by a decrease of 1 in the flow on x<sub>35</sub>.

&emsp;&emsp;Therefore, x<sub>34</sub>, x<sub>25</sub>, and x<sub>12</sub> increase by 1, and x<sub>35</sub>, x<sub>22</sub>, and x<sub>24</sub> decrease by 1. The variable x<sub>34</sub> enters the basis and x<sub>35</sub> leaves the basis.

&emsp;&emsp;When we now return to Part I of the algorithm, we can select any basic cell. There are some small computational savings to be obtained if we choose the basic variable that just entered. Consider the new basic network tree in **Figure 3.10**. In the figure, we have reordered the warehouse and customer numbers to eliminate crossing lines.

&emsp;&emsp;Notice the edge corresponding to c′<sub>34</sub> = −6. In order to get c′<sub>34</sub> = 0, we must decrease either u<sub>3</sub> or v<sub>4</sub>. Suppose we decrease v<sub>4</sub> by 6. Then, in order to keep all other c′<sub>ij</sub> = 0 for basic edges, we must increase u<sub>1</sub> and u<sub>2</sub> by 6, and decrease v<sub>2</sub> and v<sub>5</sub> by 6. The new reduced cost matrix is shown in **Table 3.12**.

&emsp;&emsp;The total cost is ＄6 lower for a total of ＄604 in the original problem. Moreover, all of the reduced costs are now non-negative. Just as in the Simplex method, when the reduced costs are all non-negative, the current solution must be optimal.
